```python
import re
import pandas as pd
import html
import string

# load data
df_recipes = pd.read_csv('recipes.csv')

# function to remove specific patterns from strings
def remove_pattern(value):
    if isinstance(value, str):
        # Define the pattern to remove
        pattern = r'c\("https://img\.sndimg\.com/food/image/upload/\)'
        # Remove the pattern from the string
        return re.sub(pattern, '', value)
    return value

# function to unescape HTML entities
def unescape_html(value):
    if isinstance(value, str):
        return html.unescape(value)
    return value

# function to gently remove punctuation from string values
def remove_punctuation(value):
    if isinstance(value, str):
        # Define a translation table to remove punctuation
        translation_table = str.maketrans('', '', string.punctuation)
        return value.translate(translation_table)
    return value

# ffunction to remove leading 'c' from ingredient parts
def remove_leading_c(value):
    if isinstance(value, str):
        # Remove leading 'c' followed by a letter (to catch cases like 'cblueberries')
        return re.sub(r'\bc(\w)', r'\1', value)
    return value

# apply all transformations to the relevant text columns
text_columns = ['Name', 'Description', 'RecipeInstructions', 'Keywords', 'RecipeIngredientParts']
for column in text_columns:
    df_recipes[column] = df_recipes[column].apply(remove_pattern)
    df_recipes[column] = df_recipes[column].apply(remove_punctuation)
    df_recipes[column] = df_recipes[column].apply(unescape_html)
    df_recipes[column] = df_recipes[column].apply(remove_leading_c)

# function to convert ISO 8601 duration strings to total minutes
def iso8601_to_minutes(duration):
    if isinstance(duration, str):
        match = re.match(r'PT(?:(\d+)H)?(?:(\d+)M)?', duration)
        if match:
            hours = int(match.group(1)) if match.group(1) else 0
            minutes = int(match.group(2)) if match.group(2) else 0
            return hours * 60 + minutes
    return 0  # Default value if the duration is not a valid string

# convert time columns
time_columns = ['CookTime', 'PrepTime', 'TotalTime']
for column in time_columns:
    df_recipes[column] = df_recipes[column].apply(iso8601_to_minutes)

# convert all column names to lowercase
df_recipes.columns = df_recipes.columns.str.lower()

# apply lowercase to all string data in the DataFrame
df_recipes = df_recipes.applymap(lambda s: s.lower() if isinstance(s, str) else s)

# filter the DataFrame
columns_to_keep = [
    'recipeid',
    'name',
    'cooktime',
    'preptime',
    'totaltime',
    'recipecategory',
    'recipeingredientparts',
    'aggregatedrating',
    'reviewcount',
    'recipeinstructions'
]

df_recipes_filtered = df_recipes[columns_to_keep]

# handle missing values
df_recipes_filtered.dropna(subset=['name', 'recipecategory', 'recipeinstructions'], inplace=True)
df_recipes_filtered['aggregatedrating'].fillna(0, inplace=True)
df_recipes_filtered['reviewcount'].fillna(0, inplace=True)
df_recipes_filtered['recipeingredientparts'].fillna('', inplace=True)

# ensure correct data types
df_recipes_filtered['recipeid'] = df_recipes_filtered['recipeid'].astype(int)
df_recipes_filtered['aggregatedrating'] = df_recipes_filtered['aggregatedrating'].astype(float)
df_recipes_filtered['reviewcount'] = df_recipes_filtered['reviewcount'].astype(float)

# function to remove special characters
def remove_special_characters(text):
    return re.sub(r'[^A-Za-z0-9\s]', '', text)

# apply the function to all string columns
for column in df_recipes_filtered.select_dtypes(include=['object']).columns:
    df_recipes_filtered[column] = df_recipes_filtered[column].apply(remove_special_characters)

# save the processed DataFrame to a new CSV file
df_recipes_filtered.to_csv('recipes_filtered.csv', index=False)

# display the DataFrame
display(df_recipes_filtered.head())

# print the shapes of the original and filtered DataFrames
print("Original DataFrame shape:", df_recipes.shape)
print("Filtered DataFrame shape:", df_recipes_filtered.shape)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# load the processed data
df_recipes_filtered = pd.read_csv('recipes_filtered.csv')

# function to log transform review counts
def log_transform_review_count(df):
    df['log_reviewcount'] = np.log1p(df['reviewcount'])
    return df

# function to remove outliers from review counts
def remove_review_count_outliers(df, quantile=0.99):
    threshold = df['reviewcount'].quantile(quantile)
    return df[df['reviewcount'] <= threshold]

# apply the log transformation
df_recipes_filtered = log_transform_review_count(df_recipes_filtered)

# remove outliers from review counts
df_recipes_no_outliers = remove_review_count_outliers(df_recipes_filtered)

# plot the log-transformed review count distribution
plt.figure(figsize=(12, 6))
sns.histplot(df_recipes_filtered['log_reviewcount'], bins=20, kde=True)
plt.title('Distribution of Log Transformed Review Counts')
plt.xlabel('Log Review Count')
plt.ylabel('Frequency')
plt.show()

# plot the review count distribution without outliers
plt.figure(figsize=(12, 6))
sns.histplot(df_recipes_no_outliers['reviewcount'], bins=20, kde=True)
plt.title('Distribution of Review Counts (Without Outliers)')
plt.xlabel('Review Count')
plt.ylabel('Frequency')
plt.show()
```

```python
import pandas as pd
import re

# load the dataset
df_reviews = pd.read_csv('reviews.csv')

# lowercase column names
df_reviews.columns = df_reviews.columns.str.lower()

# drop rows with NaNs in essential columns"
df_reviews.dropna(inplace=True)

# drop duplicates
df_reviews = df_reviews.drop_duplicates()

# drop unnecessary columns
df_reviews = df_reviews.drop(columns=['datesubmitted', 'datemodified', 'authorid', 'authorname', 'unnamed: 0', 'review'])

# handle missing values
df_reviews['reviewid'] = df_reviews['reviewid'].astype(int)
df_reviews['recipeid'] = df_reviews['recipeid'].astype(int)
df_reviews['rating'] = df_reviews['rating'].astype(float)

# function to remove special characters
def remove_special_characters(text):
    if isinstance(text, str):
        return re.sub(r'[^A-Za-z0-9\s]', '', text)
    return text

# convert all object type columns to strings and apply the function
for column in df_reviews.select_dtypes(include=['object']).columns:
    df_reviews[column] = df_reviews[column].astype(str).apply(remove_special_characters)

# save the cleaned and sampled dataset
df_reviews.to_csv('modif_reviews.csv', index=False)

# display the sampled dataframe
print(df_reviews)
```
